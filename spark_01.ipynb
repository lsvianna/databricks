## Base do experimento

```python
# Sintetiza uma base de dados industrial
df = (
    spark.range(1, 50001)
    .withColumnRenamed('id', 'id_leitura')
    .withColumn('maquina_id', expr("concat('M', lpad(cast((id_leitura % 8) + 1 as string), 2, '0'))"))
    .withColumn(
        'linha',
        expr("""
            CASE
                WHEN id_leitura % 3 = 0 THEN 'Linha A'
                WHEN id_leitura % 3 = 1 THEN 'Linha B'
                ELSE 'Linha C'
            END
        """)
    )
    .withColumn(
        'turno',
        expr("""
            CASE
                WHEN id_leitura % 2 = 0 THEN 'Diurno'
                ELSE 'Noturno'
            END
        """)
    )
    .withColumn('temperatura', round(60 + rand(7) * 35, 2))
    .withColumn('vibracao', round(2 + rand(11) * 10, 2))
    .withColumn('consumo_energia', round(80 + rand(19) * 70, 2))
    .withColumn('pecas_produzidas', expr('CAST(80 + rand(23) * 120 AS INT)'))
    .withColumn('pecas_refugadas', expr('CAST(rand(29) * 8 AS INT)'))
    .withColumn(
        'status_operacao',
        when((col('temperatura') > 85) | (col('vibracao') > 9), 'alerta').otherwise('normal')
    )
)
```
