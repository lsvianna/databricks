### Cria uma query com filtros e agrupamentos

````python
%sql
SELECT
  pickup_zip,
  COUNT(*) AS total_corridas,
  ROUND(AVG(trip_distance), 2) AS distancia_media,
  ROUND(AVG(fare_amount), 2) AS tarifa_media
FROM tabela_view
GROUP BY pickup_zip
ORDER BY total_corridas DESC
LIMIT 10
````

# Atividade Prática

Construa uma consulta SQL para visualizar os **5 locais de destino (`dropoff_zip`) com mais corridas**.

## O que sua query deve ter:

```sql
SELECT
  dropoff_zip,
  COUNT(*) AS total_corridas
FROM tabela_view
GROUP BY ___________
ORDER BY ___________ DESC
LIMIT 5
```

## Após executar:
1. Clique no botão **"+"** acima da tabela de resultados
2. Escolha **"Visualization"**
3. Selecione um gráfico de **barras**
4. Observe: qual região tem mais corridas?

````python
%sql
SELECT
  dropoff_zip,
  COUNT(*) AS total_corridas
FROM tabela_view
WHERE dropoff_zip IS NOT NULL
GROUP BY dropoff_zip
ORDER BY total_corridas DESC
LIMIT 5
````
